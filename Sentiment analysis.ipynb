```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
from sklearn.model_selection import cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
import string
punc=string.punctuation
ps=PorterStemmer()
lm=WordNetLemmatizer()
def clean_text(text):
    sw=stopwords.words('english')
    sw.remove('not')#enter all like-> not, did not ,have not....
    sw.remove('won')
    new_text=re.sub(f'[{punc}]','',text)
    words=word_tokenize(new_text)
    new_words=[]
    for w in words:
        if(w not in sw):
            nw=ps.stem(w)
            new_words.append(nw)#stemming
    return ' '.join(new_words)
```

```python
def review_analysis(text,frm):
    df=pd.read_csv('D:/python home/ML 2.0/Pandas/dataset/dataset/sentiment/Restaurant_Reviews.txt',delimiter='\t')
    df['Review']=df.Review.map(clean_text)
    tv=TfidfVectorizer()
    X=tv.fit_transform(df.Review).toarray()
    y=df.Liked
    from sklearn.naive_bayes import MultinomialNB
    #print(cross_val_score(MultinomialNB(),X,y,cv=10).mean())
    mnb=MultinomialNB()
    mnb.fit(X,y)
    test_sample=text
    X_test=tv.transform([clean_text(test_sample)]).toarray()
    result=mnb.predict(X_test)
    if(result==1):
        Label(frm,text='Positive Review',font=("",20,'bold'),bg='blue',fg='#7FFF00').place(x=560,y=400)
    else:
        Label(frm,text='Negative Review',font=("",20,'bold'),bg='blue',fg='#7FFF00').place(x=560,y=400)
```

```python
def spam_detection(text,frm):
    df=pd.read_csv('D:/python home/ML 2.0/Pandas/dataset/dataset/sentiment/spam_ham.txt',delimiter='\t',header=None,names=['states','msg'])
    df['msg']=df.msg.map(clean_text)
    tv=TfidfVectorizer()
    X=tv.fit_transform(df.msg[:1500]).toarray()
    y=df.states[:1500]
    from sklearn.naive_bayes import BernoulliNB
    #print(cross_val_score(BernoulliNB(),X,y,cv=10).mean())
    bnb=BernoulliNB()
    bnb.fit(X,y)
    test_sample=text
    X_test=tv.transform([clean_text(test_sample)]).toarray()
    result=bnb.predict(X_test)
    #print(bnb.predict_proba(X_test))
    if(result=='ham'):
        Label(frm,text='Message is Ham',font=("",20,'bold'),bg='blue',fg='#7FFF00').place(x=560,y=400)
    else:
        Label(frm,text='Message is spam',font=("",20,'bold'),bg='blue',fg='#7FFF00').place(x=560,y=400)
```

```python
def tweet_analysis(text,frm):
    #text='i miss new moon trailer'
    df=pd.read_csv('D:/python home/ML 2.0/Machine Learning/my project/sentiment Analysis/twitter.csv',encoding="ISO-8859-1").iloc[:2000,:]
    df['SentimentText']=df.SentimentText.map(clean_text)
    tv=TfidfVectorizer()
    X=tv.fit_transform(df.SentimentText).toarray()
    y=df.Sentiment

    #print(cross_val_score(LogisticRegression(solver='liblinear'),X,y,cv=10).mean())

    lr=LogisticRegression(solver='liblinear')
    lr.fit(X,y)
    test_sample=text
    X_test=tv.transform([clean_text(test_sample)]).toarray()
    result=lr.predict(X_test)
    #print(lr.predict_proba(X_test))
    #print(result)
    if(result==0):
        Label(frm,text='Negative tweet',font=("",20,'bold'),bg='blue',fg='#7FFF00').place(x=560,y=400)
    else:
        Label(frm,text='Positive tweet',font=("",20,'bold'),bg='blue',fg='#7FFF00').place(x=560,y=400)
```
